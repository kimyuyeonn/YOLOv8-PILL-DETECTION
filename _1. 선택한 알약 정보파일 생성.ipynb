```python
# 골라놓은 알약
SEL_DRUG = ['K-001562', 'K-001901', 'K-002052', 'K-004378', 'K-005391', 'K-006192', 'K-006563', 'K-006697', 'K-007024', 'K-007348', 'K-012769', 'K-013398', 'K-013900', 'K-015710', 'K-016206', 'K-016548', 'K-016688', 'K-021771', 'K-023933', 'K-024029', 'K-024127', 'K-025629', 'K-027777', 'K-028408', 'K-031490', 'K-038104', 'K-038722', 'K-040265', 'K-041149', 'K-041768', 'K-044199', 'K-044830', 'K-044834']
len(SEL_DRUG)
```

```python
# 알약정보 불러오기
import pandas as pd

df_drug = pd.read_csv('./DRUG_INFO/이미지정보.csv')
df_drug.head(2)
```

```python
# 선택된 단일 경구약 이미지 정보만 추출
df_sel_drug = df_drug.loc[(df_drug['drug_N'].isin(SEL_DRUG)) & (df_drug['json_path'].str.contains('단일'))]
df_sel_drug = df_sel_drug.sort_values(by='imgfile') # 파일명(알약식별번호) 순으로 정렬
df_sel_drug.head(2)
```

```python
# df_sel_drug.set_index('imgfile', drop=True)
df_sel_drug = df_sel_drug.drop('Unnamed: 0', axis=1)  # 필요없는 컬럼 삭제
df_sel_drug = df_sel_drug.reset_index(drop=True)      # 숫자 인덱스 재설정
df_sel_drug.head(2)
```

```python
df_sel_drug.to_csv('./DRUG_INFO/선택한알약이미지정보.csv', index=False)
```
